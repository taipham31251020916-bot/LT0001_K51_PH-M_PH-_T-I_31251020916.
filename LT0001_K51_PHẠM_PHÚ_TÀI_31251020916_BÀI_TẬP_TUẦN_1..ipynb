```python
#23.7
import osmnx as ox
import networkx as nx
import folium
from IPython.display import display

# 1. Tải mạng lưới (Sử dụng khu vực nhỏ để tránh lỗi RAM)
place_name = "District 1, Ho Chi Minh City, Vietnam"
graph = ox.graph_from_place(place_name, network_type='drive')

# 2. Xác định điểm đầu và cuối
start_coords = (10.7797, 106.6990)
end_coords = (10.7720, 106.6980)

# Tìm node gần nhất
orig_node = ox.nearest_nodes(graph, start_coords[1], start_coords[0])
dest_node = ox.nearest_nodes(graph, end_coords[1], end_coords[0])

# 3. Tính toán lộ trình
# Dijkstra
route_dijkstra = nx.shortest_path(graph, orig_node, dest_node, weight='length', method='dijkstra')
# A*
route_astar = nx.astar_path(graph, orig_node, dest_node, weight='length')

# 4. Trực quan hóa bằng Folium (Cách thủ công bền vững nhất)
# Tạo bản đồ nền trung tâm tại điểm bắt đầu
m = folium.Map(location=start_coords, zoom_start=15)

# Hàm vẽ lộ trình lên bản đồ
def plot_route(m, graph, route, color, weight):
    nodes = []
    for node in route:
        nodes.append((graph.nodes[node]['y'], graph.nodes[node]['x']))
    folium.PolyLine(nodes, color=color, weight=weight, opacity=0.8).add_to(m)

# Vẽ lộ trình Dijkstra (Xanh dương, nét to)
plot_route(m, graph, route_dijkstra, 'blue', 6)

# Vẽ lộ trình A* (Đỏ, nét nhỏ hơn để thấy cả hai)
plot_route(m, graph, route_astar, 'red', 2)

# Thêm Marker
folium.Marker(start_coords, popup="Bắt đầu", icon=folium.Icon(color='green')).add_to(m)
folium.Marker(end_coords, popup="Kết thúc", icon=folium.Icon(color='red')).add_to(m)

display(m)
```

```python
#23.8 Phương pháp heuristic
# 1. Cài đặt và nhập thư viện
import folium
import pandas as pd
import numpy as np
from folium import plugins

# 2. Khởi tạo dữ liệu giả lập
# Tọa độ khu vực Quận 1, TP.HCM
center_lat, center_lon = 10.7769, 106.7009

# Tạo 5 khách hàng ngẫu nhiên
np.random.seed(42)
customers = pd.DataFrame({
    'id': [f'Khách {i}' for i in range(1, 6)],
    'lat': center_lat + np.random.uniform(-0.005, 0.005, 5),
    'lon': center_lon + np.random.uniform(-0.005, 0.005, 5)
})

# Tạo 8 tài xế ngẫu nhiên
drivers = pd.DataFrame({
    'id': [f'Xe {i}' for i in range(1, 9)],
    'lat': center_lat + np.random.uniform(-0.008, 0.008, 8),
    'lon': center_lon + np.random.uniform(-0.008, 0.008, 8)
})

# 3. Thuật toán Heuristic: Ghép cặp xe gần nhất (Nearest Neighbor)
assignments = []
used_drivers = set()

for _, customer in customers.iterrows():
    min_dist = float('inf')
    best_driver_idx = None

    for idx, driver in drivers.iterrows():
        if idx not in used_drivers:
            # Tính khoảng cách Euclidean đơn giản
            dist = np.sqrt((customer['lat'] - driver['lat'])**2 + (customer['lon'] - driver['lon'])**2)
            if dist < min_dist:
                min_dist = dist
                best_driver_idx = idx

    if best_driver_idx is not None:
        used_drivers.add(best_driver_idx)
        assignments.append((customer, drivers.loc[best_driver_idx]))

# 4. Trực quan hóa trên Folium
m = folium.Map(location=[center_lat, center_lon], zoom_start=15, tiles='OpenStreetMap')

# Thêm Marker cho Khách hàng (Icon người)
for _, c in customers.iterrows():
    folium.Marker(
        [c['lat'], c['lon']],
        popup=c['id'],
        icon=folium.Icon(color='red', icon='user', prefix='fa')
    ).add_to(m)

# Thêm Marker cho Tài xế (Icon xe)
for _, d in drivers.iterrows():
    color = 'blue'
    folium.Marker(
        [d['lat'], d['lon']],
        popup=d['id'],
        icon=folium.Icon(color=color, icon='car', prefix='fa')
    ).add_to(m)

# Vẽ đường nối ghép cặp (Màu xanh lá cây)
for customer, driver in assignments:
    folium.PolyLine(
        locations=[[customer['lat'], customer['lon']], [driver['lat'], driver['lon']]],
        color='green',
        weight=3,
        opacity=0.7,
        dash_array='10'
    ).add_to(m)

# Hiển thị bản đồ trong Colab
m
```

```python
#23.9
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import folium
from sklearn.metrics import silhouette_score
# 1. Giả lập 100 điểm khách hàng quanh khu vực Quận 1, Quận 3, Quận 10
np.random.seed(10)
n_customers = 100
data = {
    'lat': np.random.uniform(10.76, 10.79, n_customers),
    'lon': np.random.uniform(106.66, 106.70, n_customers)
}
df = pd.DataFrame(data)

# 2. Áp dụng K-Means (giả sử doanh nghiệp muốn đặt 4 trạm trung chuyển)
n_clusters = 4
model = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
df['cluster'] = model.fit_predict(df[['lat', 'lon']])

# Tọa độ các trạm trung chuyển (Centroids)
centroids = model.cluster_centers_

# 3. Tạo bản đồ tương tác
m = folium.Map(location=[10.775, 106.68], zoom_start=14)

# Bảng màu cho các cụm
colors = ['red', 'blue', 'green', 'orange']

# Vẽ khách hàng
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=3,
        color=colors[int(row['cluster'])],
        fill=True,
        fill_opacity=0.6
    ).add_to(m)

# Vẽ trạm trung chuyển (Icon đặc biệt)
for i, center in enumerate(centroids):
    folium.Marker(
        location=[center[0], center[1]],
        popup=f"TRẠM TỐI ƯU {i+1}",
        icon=folium.Icon(color='darkred', icon='info-sign')
    ).add_to(m)

# Hiển thị bản đồ ngay trong Colab
m
```

```python
#23.11
import pandas as pd
import numpy as np
import folium
from folium.plugins import HeatMapWithTime
from sklearn.ensemble import RandomForestRegressor
from datetime import datetime, timedelta

# Giả lập dữ liệu cho 5 khu vực trong 24 giờ
data = []
locations = {
    "Quận 1": [10.7769, 106.7009],
    "Quận 7": [10.7289, 106.7067],
    "Thủ Đức": [10.8491, 106.7733],
    "Bình Thạnh": [10.8106, 106.7091],
    "Tân Bình": [10.8015, 106.6558]
}

for hour in range(24):
    for loc_name, coords in locations.items():
        # Giả lập nhu cầu cao vào giờ cao điểm (8h và 18h)
        base_demand = 10
        if 7 <= hour <= 9 or 17 <= hour <= 19:
            base_demand = np.random.randint(50, 100)
        else:
            base_demand = np.random.randint(5, 30)

        data.append([hour, coords[0], coords[1], base_demand])

df = pd.DataFrame(data, columns=['hour', 'lat', 'lon', 'demand'])

 # Huấn luyện mô hình
X = df[['hour', 'lat', 'lon']]
y = df['demand']

model = RandomForestRegressor(n_estimators=100)
model.fit(X, y)

# Dự báo nhu cầu cho giờ tiếp theo (ví dụ giờ thứ 25 - tương đương 1h sáng hôm sau)
next_hour = 18 # Dự báo cho giờ cao điểm tối
test_loc = [10.7769, 106.7009] # Tọa độ Quận 1
predicted = model.predict([[next_hour, test_loc[0], test_loc[1]]])
print(f"Dự báo nhu cầu tại Quận 1 lúc {next_hour}h: {int(predicted[0])} đơn hàng")

# Tạo bản đồ trung tâm tại TP.HCM
m = folium.Map(location=[10.7769, 106.7009], zoom_start=12)

# Chuẩn bị dữ liệu cho HeatMapWithTime
heat_data = []
for hour in range(24):
    hour_data = df[df['hour'] == hour]
    # Folium yêu cầu định dạng: [lat, lon, weight]
    heat_data.append(hour_data[['lat', 'lon', 'demand']].values.tolist())

# Thêm hiệu ứng HeatMap theo thời gian
HeatMapWithTime(heat_data, auto_play=True, max_opacity=0.8).add_to(m)

# Hiển thị bản đồ
m
```

```python
#23.12
import folium
import numpy as np

# 1. Khởi tạo dữ liệu: Kho (Depots) và Điểm giao hàng (Customers) tại TP.HCM
depots = {
    'Kho A (Quận 1)': [10.776, 106.700],
    'Kho B (Quận 3)': [10.785, 106.685]
}

customers = [
    [10.771, 106.698], [10.785, 106.710], [10.773, 106.685],
    [10.790, 106.680], [10.765, 106.690], [10.780, 106.675]
]

# Hàm tính khoảng cách Euclidean
def get_dist(p1, p2):
    return np.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)

# 2. Thuật toán tối ưu: Phân nhóm và Sắp xếp lộ trình
routes = {name: [] for name in depots.keys()}

for c in customers:
    # Bước 1: Tìm kho gần nhất cho mỗi khách hàng (Phân vùng)
    closest_depot = min(depots.keys(), key=lambda d: get_dist(c, depots[d]))
    routes[closest_depot].append(c)

# 3. Trực quan hóa trên bản đồ
m = folium.Map(location=[10.776, 106.700], zoom_start=14)
colors = {'Kho A (Quận 1)': 'red', 'Kho B (Quận 3)': 'blue'}

for depot_name, points in routes.items():
    depot_geo = depots[depot_name]

    # Vẽ Marker cho Kho
    folium.Marker(depot_geo, popup=depot_name,
                  icon=folium.Icon(color=colors[depot_name], icon='home')).add_to(m)

    # Sắp xếp lộ trình theo "Nearest Neighbor" từ kho (Heuristic)
    current_pos = depot_geo
    unvisited = points.copy()
    ordered_route = [depot_geo]

    while unvisited:
        next_pt = min(unvisited, key=lambda p: get_dist(current_pos, p))
        ordered_route.append(next_pt)
        unvisited.remove(next_pt)

        # Vẽ Marker cho khách hàng
        folium.CircleMarker(next_pt, radius=6, color=colors[depot_name],
                            fill=True, popup="Điểm giao").add_to(m)

    ordered_route.append(depot_geo) # Quay về kho

    # Vẽ đường nối tối ưu
    folium.PolyLine(ordered_route, color=colors[depot_name], weight=4, opacity=0.7).add_to(m)

# Hiển thị bản đồ
m
```

```python
#23.14
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
import random
from IPython.display import clear_output
import time

# --- CẤU HÌNH ---
PLACE_NAME = "Ben Thanh Market, Ho Chi Minh City, Vietnam"
DIST = 1000
NUM_VEHICLES = 3
TIME_STEPS = 20

# 1. Tải và xử lý đồ thị
print("Đang tải bản đồ...")
G_raw = ox.graph_from_address(PLACE_NAME, dist=DIST, network_type='drive')

#Thử các cách chạy xem cái nào tương thích bản hiện tại
try:
    G = ox.truncate.largest_component(G_raw, strongly=True)
except AttributeError:
    try:
        G = ox.utils_graph.get_largest_component(G_raw, strongly=True)
    except:
        largest_scc = max(nx.strongly_connected_components(G_raw), key=len)
        G = G_raw.subgraph(largest_scc).copy()

nodes = list(G.nodes())

# 2. Lớp Phương tiện (Giữ nguyên logic)
class Vehicle:
    def __init__(self, v_id, start_node):
        self.v_id = v_id
        self.current_node = start_node
        self.route = []
        self.status = "IDLE"
        self.color = ['red', 'blue', 'green', 'orange'][v_id % 4]

    def move(self):
        if self.route:
            self.current_node = self.route.pop(0)
            if not self.route:
                self.status = "IDLE"

# 3. Khởi tạo xe
vehicles = [Vehicle(i, random.choice(nodes)) for i in range(NUM_VEHICLES)]

# 4. Mô phỏng:
for t in range(TIME_STEPS):
    # Order
    if random.random() > 0.4:
        order_node = random.choice(nodes)

        # Tìm xe rảnh gần nhất bằng thuật toán Dijkstra
        best_vehicle = None
        min_dist = float('inf')

        for v in vehicles:
            if v.status == "IDLE":
                try:
                    d = nx.shortest_path_length(G, v.current_node, order_node, weight='length')
                    if d < min_dist:
                        min_dist = d
                        best_vehicle = v
                except (nx.NetworkXNoPath, nx.NodeNotFound):
                    continue

        if best_vehicle:
            try:
                best_vehicle.route = nx.shortest_path(G, best_vehicle.current_node, order_node, weight='length')
                best_vehicle.status = "BUSY"
                print(f"T={t}: Đã điều phối Xe {best_vehicle.v_id} đón khách.")
            except:
                pass

    # Cập nhật vị trí
    for v in vehicles:
        v.move()

    # Trực quan hóa bản đồ
    clear_output(wait=True)
    fig, ax = ox.plot_graph(G, node_size=0, edge_color='#BDC3C7', edge_linewidth=0.8, show=False, close=False)

    for v in vehicles:
        x, y = G.nodes[v.current_node]['x'], G.nodes[v.current_node]['y']
        ax.scatter(x, y, c=v.color, s=200, edgecolors='white', label=f'Xe {v.v_id} ({v.status})', zorder=5)

    plt.title(f"Hệ thống điều phối thực tế - Step {t}")
    plt.legend(loc='upper right')
    plt.show()
    time.sleep(0.8)
```

```python
#23.15
# 1. CÀI ĐẶT THƯ VIỆN
!pip install -q ortools folium scikit-learn pandas numpy

import numpy as np
import pandas as pd
import folium
from sklearn.ensemble import RandomForestRegressor
from ortools.constraint_solver import pywrapcp, routing_enums_pb2

# 2. XÂY DỰNG MÔ HÌNH AI DỰ BÁO (PREDICTIVE MODEL)
# Giả lập dữ liệu: [Mức rác hiện tại %, Giờ trong ngày, Cuối tuần (0/1)] -> Mức rác sau 2h
X_train = np.array([
    [10, 7, 0], [40, 9, 0], [70, 11, 0], [20, 13, 1], [50, 15, 1],
    [85, 17, 0], [15, 8, 1], [60, 10, 0], [90, 19, 0], [30, 14, 1]
])
y_train = np.array([25, 60, 95, 45, 75, 100, 40, 85, 100, 55])

# Huấn luyện mô hình
ai_model = RandomForestRegressor(n_estimators=50, random_state=42)
ai_model.fit(X_train, y_train)

# 3. THIẾT LẬP DỮ LIỆU ĐỊA LÝ (GIS DATA)
# Giả lập danh sách các thùng rác quanh khu vực UEH Nguyễn Tri Phương
depot_pos = (10.7605, 106.6655) # Bãi xe rác
trash_bins = [
    {"id": "DEPOT", "pos": depot_pos, "current_stat": [0, 8, 0]}, # Bến xuất phát
    {"id": "Bin_01", "pos": (10.7620, 106.6675), "current_stat": [75, 8, 0]},
    {"id": "Bin_02", "pos": (10.7590, 106.6690), "current_stat": [20, 8, 0]},
    {"id": "Bin_03", "pos": (10.7580, 106.6640), "current_stat": [85, 8, 0]},
    {"id": "Bin_04", "pos": (10.7635, 106.6630), "current_stat": [60, 8, 0]},
    {"id": "Bin_05", "pos": (10.7650, 106.6660), "current_stat": [10, 8, 0]},
    {"id": "Bin_06", "pos": (10.7615, 106.6695), "current_stat": [90, 8, 0]},
]
# Bước này lọc ra các thùng rác được AI dự báo là sẽ đầy (> 80%)
target_locations = [trash_bins[0]] # Luôn bao gồm Depot
for b in trash_bins[1:]:
    pred = ai_model.predict([b["current_stat"]])[0]
    if pred >= 80:
        b["predicted_fill"] = pred
        target_locations.append(b)

print(f"Hệ thống xác định {len(target_locations)-1} thùng rác cần thu gom ngay lập tức.")

# 4. GIẢI BÀI TOÁN TỐI ƯU LỘ TRÌNH (VRP)
# =================================================================
def get_distance(p1, p2):
    # Tính khoảng cách Manhattan đơn giản để mô phỏng đường phố
    return int((abs(p1[0]-p2[0]) + abs(p1[1]-p2[1])) * 100000)

def solve_routing(locations):
    num_locs = len(locations)
    manager = pywrapcp.RoutingIndexManager(num_locs, 1, 0)
    routing = pywrapcp.RoutingModel(manager)

    dist_matrix = [[get_distance(locations[i]['pos'], locations[j]['pos'])
                    for j in range(num_locs)] for i in range(num_locs)]

    def distance_callback(from_index, to_index):
        return dist_matrix[manager.IndexToNode(from_index)][manager.IndexToNode(to_index)]

    transit_callback_index = routing.RegisterTransitCallback(distance_callback)
    routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)

    search_params = pywrapcp.DefaultRoutingSearchParameters()
    search_params.first_solution_strategy = routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC

    solution = routing.SolveWithParameters(search_params)

    route = []
    if solution:
        index = routing.Start(0)
        while not routing.IsEnd(index):
            route.append(manager.IndexToNode(index))
            index = solution.Value(routing.NextVar(index))
        route.append(manager.IndexToNode(index))
    return route

optimized_indices = solve_routing(target_locations)

# 5. TRỰC QUAN HÓA (VISUALIZATION)
m = folium.Map(location=depot_pos, zoom_start=16, tiles="OpenStreetMap")

# Vẽ tất cả các thùng rác hiện có
for b in trash_bins:
    pred = ai_model.predict([b["current_stat"]])[0]
    is_target = pred >= 80
    color = 'red' if is_target else 'green'
    icon = 'trash' if is_target else 'info-sign'

    folium.Marker(
        location=b["pos"],
        popup=f"ID: {b['id']}<br>Dự báo: {pred:.1f}%",
        icon=folium.Icon(color=color, icon=icon)
    ).add_to(m)

# Vẽ lộ trình xe rác đi qua các điểm "đỏ"
route_coords = [target_locations[i]['pos'] for i in optimized_indices]
folium.PolyLine(
    route_coords,
    color="#2c3e50",
    weight=5,
    opacity=0.7,
    dash_array='10'
).add_to(m)

# Thêm hiệu ứng mũi tên hướng đi (Mô phỏng)
folium.RegularPolygonMarker(location=route_coords[1], fill_color='blue', number_of_sides=3, radius=10, rotation=90).add_to(m)

display(m)
```
